```python
x=df.iloc[:,:-1]
x
```

```python
df['Quality']=df['Quality'].map({'good':0,'bad':1})
```

```python
df
```

```python
y=df.iloc[:,-1]
```

```python
#train test split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytain,ytest=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# define the multinomial logistic regression model For multiclass classification
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
xstrain=scaler.fit_transform(xtrain)
```

```python
xstest=scaler.transform(xtest)
```

```python
model.fit(xstrain,ytain)
```

```python
ypred=model.predict(xstest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(ypred,ytest))
print(accuracy_score(ypred,ytest))
print(classification_report(ypred,ytest))
```

```python
#gridsearchcv
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'penalty':('l1','l2','Elasticnet',None),'C':[1,10,20]}
clf=GridSearchCV(model,parameters,cv=5)
```

```python
clf.fit(xstrain,ytain)
```

```python
clf.best_params_
```

```python
classifier=LogisticRegression(C=1,penalty='l2')
```

```python
classifier.fit(xstrain,ytain)
```

```python
ypred=classifier.predict(xstest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(ypred,ytest))
print(accuracy_score(ypred,ytest))
print(classification_report(ypred,ytest))
```
